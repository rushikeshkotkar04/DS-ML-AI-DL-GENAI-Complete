## Converting Whole text into vectors using different methods

#### 1.Bag of words
  - Converting text into vectors 

```python
import pandas as pd
messages=pd.read_csv('E:\DS_ML_AI_DL_GENAI_practice\SMSSpamCollection',sep='\t',names=["label","message"])
```

```python
messages
```

```python
## Data cleaning and preprocessing
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(messages)):
  review=re.sub('[^a-zA-Z]',' ',messages["message"][i])
  review=review.lower()
  review=review.split()
  review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
  review=" ".join(review)
  corpus.append(review)
```

```python
corpus
```

```python
## Creating bag of words
## For enabling binary back of words we have to use binary=True in the object of CountVectorizer
## max_features means how many clms we have to make
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=100,binary=True,)
```

```python
x=cv.fit_transform(corpus).toarray()
```

```python
x.shape
```

```python
cv.vocabulary_
```

# N-Grams 
  - UniGram (1,1):it makes features of every one word of vector
  - BiGram (1,2):It makes feature for vcombination two words of vector
  - TriGram (1,3):It makes features of combination of three words of vector
  in short
  - N grams (1,n): It makes feature of combination of n words of vector

```python
from sklearn.feature_extraction.text import CountVectorizer
cv_nGrams=CountVectorizer(max_features=2500,binary=True,ngram_range=(1,2))
x=cv_nGrams.fit_transform(corpus).toarray()
```

```python
cv_nGrams.vocabulary_
```

## Implementing TF IDS

```python
import pandas as pd
messages=pd.read_csv('E:\DS_ML_AI_DL_GENAI_practice\SMSSpamCollection',sep='\t',names=["label","message"])
```

```python
messages
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
wordLemmatizer=WordNetLemmatizer()
```

```python
corpus=[]
for i in range(0,len(messages)):
  review=re.sub('[^a-zA-Z]',' ',messages["message"][i])
  review=review.lower()
  review=review.split()
  review=[wordLemmatizer.lemmatize(word) for word in review if not word in stopwords.words('english')]
  review=" ".join(review)
  corpus.append(review)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer(max_features=100,ngram_range=(1,3))
x=tfidf.fit_transform(corpus).toarray()
```

```python
tfidf.vocabulary_
```
